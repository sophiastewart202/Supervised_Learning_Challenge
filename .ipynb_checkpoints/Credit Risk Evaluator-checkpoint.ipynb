### Hypothesis
In this machine learning challenge, I compare the effectiveness of the logistic regression model and the random forest classifier at predicting high loan risk. I think the logistic regression model will do better because it works well with lots of data, which is exactly what we have in the '2019loans' csv (over 12,000 rows!). The random forest classifier also has a strong tendency to be overfit -- or in other words, memorize the training data -- which is a big downside. However, the only way to know for sure is to do the work and see what happens, so scroll down to see the results of my comparison!

```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import confusion_matrix
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# clean train_df
train_df = train_df.copy().drop('Unnamed: 0', axis=1)
train_df = train_df.drop('index', axis=1)
```

```python
# clean test_df
test_df = test_df.copy().drop('Unnamed: 0', axis=1)
test_df = test_df.drop('index', axis=1)
```

```python
# Convert categorical data to numeric and drop target data
train_data = pd.get_dummies(train_df)
X_train = train_data.copy().drop('loan_status_high_risk', axis=1)
X_train = X_train.copy().drop('loan_status_low_risk', axis=1)
X_train
```

```python
# separate target feature for training data
y_train = train_data['loan_status_high_risk']
y_train.shape
```

```python
# Convert categorical data to numeric and drop target data
test_data = pd.get_dummies(test_df)
X_test = test_data.copy().drop('loan_status_high_risk', axis=1)
X_test = X_test.copy().drop('loan_status_low_risk', axis=1)
X_test
```

```python
# separate target feature for testing data
y_test = test_data['loan_status_high_risk']
y_test.shape
```

```python
# add missing dummy variables to testing set

# create a list of conditions that determine missing dummy variables
conditions = [
    (X_test['debt_settlement_flag_N'] == 1),
    (X_test['debt_settlement_flag_N'] == 0)
    ]

# create a list of the values we want to assign for each condition
values = [0, 1]

# create a new column and use np.select to assign values to it using our lists as arguments
X_test['debt_settlement_flag_Y'] = np.select(conditions, values)

X_test
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
clf = LogisticRegression(solver='lbfgs', max_iter=150)
clf
```

```python
clf.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {clf.score(X_train, y_train)}")
print(f"Testing Data Score: {clf.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=26, n_estimators=150).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf = LogisticRegression(solver='lbfgs', max_iter=150)
clf.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {clf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {clf.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=26, n_estimators=150).fit(X_train_scaled, y_train)
```

```python
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

### Conclusion: Hypothesis verified!
Before scaling the data, neither of the models are very good. The random forest classifer is WAY overfit, but gets a better testing score, and the logistic regression model has a much better fit, but a correct prediction rate scarcely better than guessing. 
After scaling the data, the logistic regression models keeps a good fit as well as vastly improves the rate at which it predicts high loan risk correctly -- about 25% higher than before scaling and 10% higher than the random forest classifier after scaling -- making the logistic regression model the clear winner in this case.
